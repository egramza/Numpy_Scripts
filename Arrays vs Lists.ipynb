Arrays vs List

```python
import numpy as np
```

```python
L = [1, 2, 3]
A = np.array([1, 2, 3])
```

```python
for e in L:
  print(e)
```

```python
for e in A:
  print(e)
```

```python
L.append(4)
```

```python
L
```

```python
A.append(4)
```

```python
L + [5]
```

```python
A + np.array([4])
```

```python
A + np.array([4, 5, 6])
```

```python
A + np.array([4, 5])
```

```python
2 * A
```

```python
2 * L
```

```python
L2 = []
for e in L:
  L2.append(e + 3)
```

```python
L2
```

```python
L2 = [e + 3 for e in L]
L2
```
